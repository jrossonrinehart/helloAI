```python
import numpy as np
import pandas as pd
import tensorflow_decision_forests as tfdf
```

```python
pathname = "/content/drive/MyDrive/AI files/car_evaluation.csv"

cars = pd.read_csv(pathname)
cars.tail(10)
```

```python
cars = cars.rename(columns={'vhigh':'buyingPrice', 'vhigh.1':'maintenanceCost', '2':'numDoors', '2.1': 'numSeats','small':'lugboot','low':'safety', 'unacc':'acceptability'})
```

```python
label = "acceptability"

classes = list(cars[label].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars[label] = cars[label].map(classes.index)

```

```python
classes = list(cars['buyingPrice'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['buyingPrice'] = cars['buyingPrice'].map(classes.index)
```

```python
classes = list(cars['maintenanceCost'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['maintenanceCost'] = cars['maintenanceCost'].map(classes.index)
```

```python
classes = list(cars['numDoors'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['numDoors'] = cars['numDoors'].map(classes.index)
```

```python
classes = list(cars['numSeats'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['numSeats'] = cars['numSeats'].map(classes.index)
```

```python
classes = list(cars['lugboot'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['lugboot'] = cars['lugboot'].map(classes.index)
```

```python
classes = list(cars['safety'].unique())
print(f"Label classes: {classes}")
# >> Label classes: ['unacc', 'acc', 'vgood', 'good']

cars['safety'] = cars['safety'].map(classes.index)
```

```python
np.random.seed(1)
# Use the ~10% of the examples as the testing set
# and the remaining ~90% of the examples as the training set.
test_indices = np.random.rand(len(cars)) < 0.1
cars_train = cars[~test_indices]
cars_test = cars[test_indices]

print("Training examples: ", len(cars_train))
# >> Training examples: 1554

print("Testing examples: ", len(cars_test))
# >> Testing examples: 173
```

```python
cars_train.head(30)
```

```python
tf_train_dataset = tfdf.keras.pd_dataframe_to_tf_dataset(cars_train, label=label)

model = tfdf.keras.CartModel()
model.fit(tf_train_dataset)
```

```python
tfdf.model_plotter.plot_model_in_colab(model, max_depth=10)
```

```python
model.compile("accuracy")
print("Train evaluation: ", model.evaluate(tf_train_dataset, return_dict=True))
# >> Train evaluation:  {'loss': 0.0, 'accuracy': 0.943}

tf_test_dataset = tfdf.keras.pd_dataframe_to_tf_dataset(cars_test, label=label)
print("Test evaluation: ", model.evaluate(tf_test_dataset, return_dict=True))
# >> Test evaluation:  {'loss': 0.0, 'accuracy': 0.948}
```

```python

outputs = model.predict(tf_train_dataset)
answers = [np.argmax(output) for output in outputs]

#create two pandas series and name index for answer_df
answers_df = pd.Series(answers)

#training data labels
labels_df = cars_train['acceptability']

#this just renames column headers
answers_df = answers_df.rename("prediction")
labels_df = labels_df.rename('true_value')

#create a dataframe
table = pd.concat([answers_df, labels_df], axis=1)
table.shape

#create new column that compares predictions to labels ('acceptability)
table['correct'] = table['prediction'] == table['true_value']

#table for only false predictions
negatives = table[table['correct']==False]

negatives.head(50)
```
