```python
import pandas as pd# list for column headers

df = pd.read_csv("/content/drive/MyDrive/AI files/mushroom.csv")
print(df.shape)
# print head of data set
print(df.head())
```

```python
df.head(10)
```

```python
for column in df:
  df[column] = df[column].str[1:]
```

```python
for column in df:
  df[column] = df[column].str.replace("'", '')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df = pd.get_dummies(df, columns=['cap-shape', 'cap-surface', 'cap-color', 'bruises%3F', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'])

colvals = {'p':1, 'e': 0}
df = df.replace({'class':colvals})
```

```python
features = df.drop('class', axis=1)
labels = df['class']
```

```python
from sklearn.model_selection import train_test_split# implementing train-test-split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.33, random_state=66)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

# random forest model creation
rfc = RandomForestClassifier()

rfc.fit(X_train,y_train)
# predictions
rfc_predict = rfc.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rfc_cv_score = cross_val_score(rfc, features, labels, cv=10, scoring='roc_auc')
```

#I don't know why the scoring is 100%

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, rfc_predict))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, rfc_predict))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```
